1\. In which direction does the wave $\mathbf{E} = E_0 \mathbf{y} (t + z/c)$ propagate?


Solution:  Since (t + z/c) remains constant as t increases only if z/c decreases correspondingly, the
wave is propagating in the -z direction.

3\. A $\lambda/2$ dipole situated with its center at the origin radiates a time-averaged power of $600 \,\rm{W}$ at a frequency of $300 \,\rm{MHz}$. A second $\lambda/2$ dipole is placed with its center at a point $P(r,\theta,\phi)$, where $r = 200 \,\rm{m},\theta = 90^{\circ} ,\phi = 40^{\circ}$. It is
oriented so that its axis is parallel to that of the transmitting antenna. What is the available power at the terminals of the second (receiving) dipole?


<!--\n
    Source: Antenna Theory: Analysis and Design, 3rd Edition, Problem 4.30, Constantine A. Balanis
-->


Solution: 
\begin{align}
\lambda & = \frac{c}{f}\\
& = \frac{3\times 10^8}{300\times 10^6}\\
& = 1\,\rm{m}\\
D_\rm{r} & = D_\rm{t} = 1.643\quad \text{for half-wavelength dipole with $\theta = 90^{\circ} $}\\
P_\rm{r} & = P_\rm{t} \left(\frac{\lambda}{4\pi R}\right)^2 D_\rm{r} D_\rm{t}\\
& = 600 \times \left(\frac{1}{4\pi \times 200}\right)^2 \times 1.643 \times 1.643 \\
& = 0.256\,\rm{mW}
\end{align}



4\. A half-wave dipole is radiating into free-space. The coordinate system is defined so that the origin is at the center of the dipole and the z-axis is aligned with the dipole. Input power to the dipole is 100 W. Assuming an overall efficiency of 50%, find the power density (in W/m 2 ) at $r = 500\,\rm{m},\theta = 60^{\circ} ,\phi = 0^{\circ} $.


<!--\n
    Source: Antenna Theory: Analysis and Design, 3rd Edition, Problem 4.30, Constantine A. Balanis
-->

Solution: 

\begin{align}
R & = 73 \rm{Ohms}, \quad \text{for half-wave dipole}\\
P_{\rm{rad}} & = 50\% \times 100\,\rm{W} = 50 \,\rm{W}\\
P_{\rm{rad}} & = \frac{\vert I_0\vert^2 R}{2}\\
\Rightarrow \vert I_0\vert^2 & = \frac{100}{73}\\
W_{\rm{av}} &= \frac{U}{r^2}\\
& = \eta_0 \frac{\vert I_0\vert^2}{8\pi^2 r^2} \left( \frac{\cos(\frac{\pi}{2} \cos\theta)}{\sin\theta} \right)^2\\
& = 120\pi \frac{100/73}{8\pi^2 500^2} \left( \frac{\cos(\frac{\pi}{4})}{\sqrt{3}/2} \right)^2\\
& = 1.744 \times 10^{-5} \,\rm{watts/m^2}
\end{align}

5\. What is the maximum solid angle $\Omega_B$ [steradians] over which a lossless matched antenna can
have constant gain $G_o$ = 40 dB?  

<!--\n
    Source: Electromagnetics and Applications, Example 10.3A by David Staelin
    Link: https://ocw.mit.edu/courses/electrical-engineering-and-computer-science/6-013-electromagnetics-and-applications-spring-2009/readings/
-->

Solution:

\begin{align}
D(\theta, \phi) & = G(\theta, \phi), \quad \text{for lossless matched antenna}\\
\int_{\Omega} D(\theta, \phi) d \Omega & = 4\pi\\
\Rightarrow   G_o \Omega_B & = 4\pi, \quad \text{ since the maximum gain results when all sidelobes have $G =
0$}
\end{align}
